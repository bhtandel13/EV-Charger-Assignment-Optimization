```python
#Author: Jennifer Han

# import numpy as np
import csv
import time
import pandas as pd
import geopy.distance 

# Range distances
dist_range = [1,3,5,10,25,50,100,250,500]

# Number of trips for each range as of November 2022 in CA
# https://www.bts.gov/browse-statistical-products-and-data/covid-related/distribution-trips-distance-national-state-and
trips_per_dist_range = [37449048,  33594077, 16624332, 19760388, 18256438, 6707869, 2406703, 902369, 131849]

# Proportion of trips per range of dist
M = sum(trips_per_dist_range)
prop_trip_per_dist = [sum(trips_per_dist_range[i:])/M for i in range(len(dist_range))]

#file = r"D:\Berkeley\2. SPRING\CE295 - Data Science for Energy\Project\month and annual average daily traffic alameda county.xlsx - Feuil1.csv"
file = r"C:\Users\Bhumi\Documents\MATLAB\CE 295\Project\month and annual average daily traffic alameda county.xlsx - Feuil1.csv"
data = pd.read_csv(file)

def traffic(x,y):

    # Sort the different PMs by ascending distance
    dist_to_PM = []
    for k in range(len(data['Latitude'])):
        dist_to_PM.append(geopy.distance.geodesic((x,y), (data['Latitude'][k], data['Longitude'][k])).miles)
    data['Distance'] = dist_to_PM
    data.sort_values('Distance', ascending = True)
    
    # Weights for the nodes
    w = []
    for index_pm in range(np.shape(data)[0]): # For each PM
        i = 0 # Index of the distance range
        if sum(w)<1:
            d = data.iloc[index_pm]['Distance']
            while d >= dist_range[i] and (i<len(dist_range)-1):
                i += 1
            if sum(w)+prop_trip_per_dist[i]<=1:
                w = np.append(w, prop_trip_per_dist[i])
            else:
                w = np.append(w, 1-sum(w))
        else:
            w = np.append(w, 0)
    print(w)
    print(np.average(data['AVERAGE BACK AND AHEAD PEAK - Annual Average Daily Traffic'], weights=w))
    return np.average(data['AVERAGE BACK AND AHEAD PEAK - Annual Average Daily Traffic'], weights=w)
    
```
